```python
x = df[['area']].values
y = df['price_usd'].values
```

### 1. Your implementation

```python
model_1 = LinearRegressionMy(learning_rate=0.0000001, n_iter=10000)
model_1.fit(x,y)
```

```python
model_1.coef_
```
